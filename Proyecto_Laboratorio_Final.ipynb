**Gráfica de las causas más frecuentes de deforestación:**

```python
dos_causas_data = df[df['CAUSA'].isin(dos_causas)]

frecuencia_por_año = dos_causas_data.groupby(['AÑO', 'CAUSA']).size().unstack() #Agrupar por AÑO y CAUSA, y contar las ocurrencias

ax = frecuencia_por_año.plot(kind='bar', figsize=(12, 6), colormap=plt.cm.colors.ListedColormap(green_palette))

umbral = 50
for p in ax.patches:
    if p.get_height() > umbral:
        ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                    ha='center', va='center', xytext=(0, 8), textcoords='offset points', rotation=0)

plt.xticks(rotation=0)
plt.title('Frecuencia de las dos causas más comunes por año')
plt.xlabel('Año')
plt.ylabel('Frecuencia')
plt.legend(title='Causa')
plt.show()
```

**Gráfica de frecuencias dos primeras causas de deforestación:**
- Al hacer la gráfica de con las dos principales causas de la deforestación en el departamento de Chocó, se evidencia que el cultivo se ha convertido, recientemente, en la causa por la que existe mayor deforestación en el departamento.
- Es interesante preguntarse y analizar que ocurrió en los años 2015 y 2019, años en los que disminuye y aumenta, respectivamente, de manera significativa la deforestación.

Se evidencia que el cultivo, la minería y los incendios son los factores que han causado mayor deforestación en el departamento.
- Es interesante ver que para el 2016, fueron los incendios a los que se les atribuye un mayor número de hectáreas deforestadas, superando las 20.000 hectáreas. Este caso es interesante dado que, en la gráfica que muestra la frecuencia de las causas de deforestación, se evidencia que el cultivo es el factor al que más se le atribuye la deforestación, sin embargo, acá se puede ver que, aunque sea el cultivo la causa más frecuente, es el incendio el factor que provocó la deforestación de un mayor número de hectáreas, seguido de la minería y el cultivo. En ese sentido, se puede afirmar que, aunque se presentaron pocos incendios, estos fueron bastante extensos y perjucidiciales para el departamento, mientras que, auqnue el cultivo es la causa más frecuente, el número de hectáreas deforestadas por este motivo es bastante baja (1.582 ha) en comparación con los incendios (20.845 ha).
